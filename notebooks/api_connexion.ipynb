```python
import requests
```

# Test the RTE API call

/!\ Be careful to delete the credentials before closing this notebook

```python
# Set the base URL
BASE_URL = "https://digital.iservices.rte-france.com"

# Set the ressources
RESSOURCE_1 = "/token/oauth/"
RESSOURCE_2 = "/open_api/actual_generation/v1/actual_generations_per_production_type"

# Content type for the headers of the Step 1
content_type = "application/x-www-form-urlencoded"

# Set my client id and secret in base 64
client_secret = ""
```

```python
###################################
# Step 1: collect the OAuth token #
###################################

# Construct the url
url = "{}{}".format(BASE_URL, RESSOURCE_1)

# Construct the headers dict
headers = {"Authorization": f"Basic {client_secret}",
           "content-type": content_type}

# Query the API with the post verb
response = requests.post(url, headers = headers)

# Show the content of the response
display(response.json())

# Store the access token
access_token = response.json().get('access_token')
```

```python
###########################################
# Step 2: query the actual generation API #
###########################################

# Construct the url
url = "{}{}".format(BASE_URL, RESSOURCE_2)

# Construct the headers dict
headers = {"Authorization": f"Bearer {access_token}"}

# Query the API with the get verb
response = requests.get(url, headers = headers)

# Show the content of the response
display(response.json())
```
